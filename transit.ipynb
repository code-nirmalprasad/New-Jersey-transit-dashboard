```python
import pandas as pd
import numpy as np
import time
import datetime
import os
```

```python
dir=os.getcwd()
for i, df in enumerate(os.listdir('data')):
    if i == 0:
        data = pd.read_csv(dir+'\\data\\'+df)
    else:
        data_temp = pd.read_csv(dir+'\\data\\'+df)
        data = pd.concat([data, data_temp], axis=0, ignore_index=True)
```

```python
data.head()
```

```python
filt1 = data['type']=='NJ Transit'
data = data.loc[filt1]
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data['scheduled_time']=pd.to_datetime(data['scheduled_time'])
scheduled_time_unix=[]
for x in data['scheduled_time']:
    scheduled_time_unix.append(datetime.datetime.timestamp(x))
data['scheduled_time_unix']=scheduled_time_unix
```

```python
data['actual_time']=pd.to_datetime(data['actual_time'])
actual_time_unix=[]
for x in data['actual_time']:
    actual_time_unix.append(datetime.datetime.timestamp(x))
data['actual_time_unix']=actual_time_unix
```

```python
delay_s=[]
for x, y in zip(data['scheduled_time_unix'], data['actual_time_unix']):
    z = y - x
    if z < 0:
        delay_s.append(0)
    else:
        delay_s.append(z)
data['delay_s'] = delay_s
```

```python
headers = list(data.columns)
```

```python
headers.remove('from')
headers.remove('to')
headers.remove('delay_minutes')
```

```python
data2 = data[headers]
data2.to_csv('transit1.csv')
```
